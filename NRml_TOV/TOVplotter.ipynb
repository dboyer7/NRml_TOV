## Straight Data

This is the straight data that comes out of the NRml Solver

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

rbar = []
rho = []
rhob = []
Pressure = []
Mass = []

with open('ooData.txt') as f: #Straight NRml Data
    reader = csv.reader(f, delimiter=' ')
    for row in reader:
        rbar.append(float(row[0]))
        rho.append(float(row[1]))
        rhob.append(float(row[2]))
        Pressure.append(float(row[3]))
        Mass.append(float(row[4]))

print("Surface at (Non-Normalized) rbar = ", rbar[-1])

fig, ax = plt.subplots()
ax.set_xlabel('rbar')
ax.set_ylabel('Solution in Corresponding units')
ax.set_title('Odie/NRml Straight Solution')

ax.plot(rbar,rho,label='rho')
ax.plot(rbar,rhob,label='rho_b')
ax.plot(rbar,Pressure,label='Press')
ax.plot(rbar,Mass,label='Mass')

fig.set_size_inches(9,9)
ax.set_yscale("log")
ax.legend()
plt.show()
```
